Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h1>Straight Push Exercise:</h1>
<font size="5"> It's expected to Change Ax (move forward and backward). Check Gx and Ay to verify whether the wrist is perpendicular to table and shakeness</font> 

<font size="3"> Read Straight_push Data from 3 users, Alan, Tae and Irfan</font> 

```python
Alan_ST = pd.read_csv("Straight_Alan.csv");
Tae_ST = pd.read_csv ("Straight_Tae.csv");
Irfan_ST = pd.read_csv("Straight_Irfan.csv");
```

```python
nan = float("NaN");
Alan_ST.replace("",nan,inplace=True)
Alan_ST.dropna(subset = ["gX"], inplace=True)
Tae_ST.replace("",nan,inplace=True)
Tae_ST.dropna(subset = ["gX"], inplace=True)
Irfan_ST.replace("",nan,inplace=True)
Irfan_ST.dropna(subset = ["gX"], inplace=True)
```

```python
plt.boxplot([Alan_ST.gX, Tae_ST.gX, Irfan_ST.gX])
plt.xticks([1,2,3], ['Alan', 'Tae', 'Irfan'])
plt.title('Gx for Straight Push')
plt.show()
```

```python
plt.boxplot([Alan_ST.aY, Tae_ST.aY, Irfan_ST.aY])
plt.xticks([1,2,3], ['Alan', 'Tae', 'Irfan'])
plt.title('Ay for Straight Push')
plt.show()
```

```python
plt.boxplot([Alan_ST.aX, Tae_ST.aX, Irfan_ST.aX])
plt.xticks([1,2,3], ['Alan', 'Tae', 'Irfan'])
plt.title('Ax for Straight Push')
plt.show()
```

```python
plt.boxplot([Alan_ST.aX, Alan_ST.aY, Alan_ST.aZ, Alan_ST.gX, Alan_ST.gY, Alan_ST.gZ, Alan_ST.Roll, Alan_ST.Pitch])
plt.xticks([1, 2, 3, 4, 5, 6, 7, 8], ['aX', 'aY', 'aZ', 'gX', 'gY', 'gZ', 'Roll', 'Pitch'])
plt.show()
```
