```python
import pandas as pd
import numpy as np
import csv   
```

<font size="5"> Basically, it can be divided into 5 steps </font>
* Read csv files with folder directory (In this csv file, it should be format like Straight.scv I uploaded to trello -> substitute column names into space (index indicator), including the last column of your csv file)
* Find the index with column names
* Write for loop -> In each iteration, it will start the new sequence and calucalte the features inside this sequence
* Create the dataframe to contain new features and label
* Write to csv file (when index is equal to zero, we subtitute column name first and data)(When index is larger than zero, it only write data)

```python
data = pd.read_csv("/Users/alan/Downloads/Bicep_FL.csv")
```

```python
data[data['aX']=='aX'].index.values
```

```python
index = data[data['aX']=='aX'].index.values;
old_idx = 0;
#path="/Users/alan/Documents/Imperial Lab/AML Group Project/Arm Guidance/" #Specify your path!

for i in index:
    sequence = data[old_idx:i].astype(float) # Change string to float
    old_idx = i+1
    ax_min = sequence['aX'].min()
    ax_max = sequence['aX'].max()
    ax_mea = sequence['aX'].mean()
    ax_std = sequence['aX'].std()
    ax_med = sequence['aX'].median()
    ay_min = sequence['aY'].min()
    ay_max = sequence['aY'].max()
    ay_mea = sequence['aY'].mean()
    ay_std = sequence['aY'].std()
    ay_med = sequence['aX'].median()
    az_min = sequence['aZ'].min()
    az_max = sequence['aZ'].max()
    az_mea = sequence['aZ'].mean()
    az_std = sequence['aZ'].std() 
    az_med = sequence['aZ'].median()    
    gx_min = sequence['gX'].min()
    gx_max = sequence['gX'].max()
    gx_mea = sequence['gX'].mean()
    gx_std = sequence['gX'].std()
    gx_med = sequence['gX'].median()    
    gy_min = sequence['gY'].min()
    gy_max = sequence['gY'].max()
    gy_mea = sequence['gY'].mean()
    gy_std = sequence['gY'].std()
    gy_med = sequence['gY'].median() 
    gz_min = sequence['gZ'].min()
    gz_max = sequence['gZ'].max()
    gz_mea = sequence['gZ'].mean()
    gz_std = sequence['gZ'].std()
    gz_med = sequence['gZ'].median() 
    pitch_min = sequence['Pitch'].min()
    pitch_max = sequence['Pitch'].max()
    pitch_mea = sequence['Pitch'].mean()
    pitch_std = sequence['Pitch'].std() 
    pitch_med = sequence['Pitch'].median() 
    roll_min = sequence['Roll'].min()
    roll_max = sequence['Roll'].max()
    roll_mea = sequence['Roll'].mean()
    roll_std = sequence['Roll'].std() 
    roll_med = sequence['Roll'].median() 
    label = 'Bicep_FL' # Change label name for each exercise
    new = pd.DataFrame([[label,ax_min,ax_max,ax_mea,ax_std,ax_med,ay_min,ay_max,ay_mea,ay_std,ay_med,az_min,az_max,az_mea,az_std,az_med,\
                         gx_min,gx_max,gx_mea,gx_std,gx_med,gy_min,gy_max,gy_mea,gy_std,gy_med,gz_min,gz_max,gz_mea,gz_std,gz_med,\
                         pitch_min,pitch_max,pitch_mea,pitch_std,pitch_med,roll_min,roll_max,roll_mea,roll_std,roll_med]],\
                         columns=['Exercise','Ax_Min','Ax_Max','Ax_Mean','Ax_std','Ax_med','Ay_Min','Ay_Max','Ay_Mean','Ay_std','Ay_med',\
                                 'Az_Min','Az_Max','Az_Mean','Az_std','Az_med',\
                                 'Gx_Min','Gx_Max','Gx_Mean','Gx_std','Gx_med','Gy_Min','Gy_Max','Gy_Mean','Gy_std','Gy_med',\
                                 'Gz_Min','Gz_Max','Gz_Mean','Gz_std','Gz_med',\
                                 'Pitch_Min','Pitch_Max','Pitch_Mean','Pitch_std','Pitch_med','Roll_Min','Roll_Max','Roll_Mean','Roll_std','Roll_med'])
    new_d = [label,ax_min,ax_max,ax_mea,ax_std,ax_med,ay_min,ay_max,ay_mea,ay_std,ay_med,az_min,az_max,az_mea,az_std,az_med,\
                         gx_min,gx_max,gx_mea,gx_std,gx_med,gy_min,gy_max,gy_mea,gy_std,gy_med,gz_min,gz_max,gz_mea,gz_std,gz_med,\
                         pitch_min,pitch_max,pitch_mea,pitch_std,pitch_med,roll_min,roll_max,roll_mea,roll_std,roll_med]
    if i == index[0]:
        with open('feature.csv','a') as fd:
            writer = csv.writer(fd)
            #writer.writerow(new)
            writer.writerow(new_d)
    else:
        with open('feature.csv','a') as fd:
            writer = csv.writer(fd)
            writer.writerow(new_d)
```
